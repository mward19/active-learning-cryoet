```python
import pandas as pd
```

```python
dataset_info = pd.read_csv('./all_datasets.csv')
dataset_info['Dataset ID'] = dataset_info['Dataset ID'].astype(str)
dataset_info.sample(15)
```

```python
print(f'Total tomograms: {dataset_info["Num Tomograms"].sum()}')
```

```python
target = 'bdello'
val_mask = dataset_info['Dataset Name'].str.contains(target, case=False)
print(f'Tomograms containing "{target}" (validation): {dataset_info[val_mask]["Num Tomograms"].sum()}')
```

I'll make those the validation set.

```python
train_mask = ~(dataset_info['Dataset Name'].str.contains(target, case=False))
print(f'Tomograms not containing "{target}" (training): {dataset_info[train_mask]["Num Tomograms"].sum()}')
```

Write to data_info file

```python
train_datasets_path = 'train_datasets.csv'
dataset_info[train_mask].to_csv(train_datasets_path, index=False)

val_datasets_path = 'val_datasets.csv'
dataset_info[val_mask].to_csv(val_datasets_path, index=False)
```
